# Text Preprocessing

In order to find the labels of SMS, we need to classify them using Classification Algorithms. 
All the machine learning models expect the input data in some numerical format, but we have the data in text format. 

We use Bag-of-words approach to represent each unique word as a number. SO the raw messages are converted into vectors (Sequence of numbers). 

First step would be splitting the messages into group of words and later removing the punctuation and stopwords

```python
import string
```

```python
print(string.punctuation)
```

```python
print(stopwords.words('English'))
```

```python
def text_process(sms):
    '''
    split words
    Remove Punctuation
    Remove stopwords
    '''
    msg = [char for char in sms if char not in string.punctuation]
    msg = ''.join(msg)
    return [word for word in msg.split() if word.lower not in stopwords.words('English')]
```

Let us apply this method to our dataset and see the changes

```python
messages['sms'].head().apply(text_process)
```

Some techniques like Stemming and POS tagging can be applied to text data for processing. 
But it may affect data like this as SMS data comprises of abbrevations and texting language which cannot be processed correctly. 

# VECTORIZATION

Now we'll convert each message, represented as a list of tokens (lemmas) above, into a vector that machine learning models can understand.
We'll do that in three steps using the bag-of-words model:

1. Count how many times does a word occur in each message (Known as term frequency)
2. Weigh the counts, so that frequent tokens get lower weight (inverse document frequency)
3. Normalize the vectors to unit length, to abstract from the original text length (L2 norm)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bag_vectors = CountVectorizer(analyzer=text_process).fit(messages['sms'])
```

```python
type(bag_vectors)
```

Now we can use .transform on our Bag-of-Words (bow) transformed object and transform the entire DataFrame of messages.

```python
sms_bag = bag_vectors.transform(messages['sms'])
```

This is a very large matrix with number of rows as total number of text messages.
Hence there will be a lot of zeroes throughout the matrix, making it a sparse matrix

```python
#Shape of sparse matrix:
sms_bag.shape
```

```python
#Number of Non-zero occurrences:
sms_bag.nnz
```

# Using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf_transformer = TfidfTransformer().fit(sms_bag)
```

```python
sms_tfidf = tfidf_transformer.transform(sms_bag)
```

```python
sms_tfidf.shape
```

# Training a Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB().fit(sms_tfidf,messages['label'])
```

Let us check how our model performs:

```python
all_pred = spam_detect_model.predict(sms_tfidf)
all_pred
```

let us evaluate our model by calculating the accuracy and recall using Classification report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(messages['label'],all_pred))
```

# Train-Test split

In the above case, we have used the entire data for both building the model and evaluating it. 
It doesn't give the accurate evaluation, hence we split our data into train and test before building model

```python
from sklearn.model_selection import train_test_split

msg_train, msg_test, label_train, label_test = \
train_test_split(messages['sms'], messages['label'], test_size=0.2)

print(len(msg_train), len(msg_test), len(msg_train) + len(msg_test))
```

# Creating a Data pipeline

Instead of working out all the preprocessing steps again, we can use the Sklearn's 'Data pipeline'. 

This will allow us to set up all the transformations that we will do to the data for future use.

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('bow', CountVectorizer(analyzer=text_process)),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
pipeline.fit(msg_train,label_train)
```

```python
predict_sms = pipeline.predict(msg_test)
```

```python
print(classification_report(label_test,predict_sms))
```
